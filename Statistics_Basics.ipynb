1.  What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

Ans. Descriptive statistics is the branch of statistics that deals with collecting, organizing, summarizing, and presenting data in order to describe the main features of a dataset,If the marks of 40 students in a class are recorded and the average mark, highest mark, lowest mark, and a bar graph of the marks are calculated, this information describes only that class. This is descriptive statistics while Inferential statistics is the branch of statistics that deals with using sample data to make conclusions, predictions, or decisions about a larger population, If the marks of the same 40 students are used to estimate the average marks of all students in the school or to test whether boys perform better than girls, this involves drawing conclusions about a population. This is inferential statistics.

2. What is sampling in statistics? Explain the differences between random
and stratified sampling.

Ans. Sampling in statistics is the process of selecting a subset (sample) from a larger group (population) in order to study the characteristics of the entire population.
Random sampling is a method of sampling in which every member of the population has an equal and independent chance of being selected while Stratified sampling is a method of sampling in which the population is divided into homogeneous subgroups called strata, and samples are randomly selected from each stratum.

3.  Define mean, median, and mode. Explain why these measures of central
tendency are important.

Ans. The mean is the measure of central tendency defined as the sum of all observations divided by the total number of observations , The median is the measure of central tendency defined as the middle value of an ordered dataset, such that half of the observations lie below it and half lie above it , The mode is the measure of central tendency defined as the value that occurs most frequently in a dataset.
Measures of central tendency are important because they summarize a large set of data into a single representative value, making it easier to understand, compare, and interpret data.

```python
''' 5.  Implement a Python program to compute the mean, median, and mode of
a given list of numbers.

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28] '''

from statistics import mean, median, mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = mean(numbers)
median_value = median(numbers)
mode_value = mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
''' 6.  Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60] '''

import statistics
import math

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

mean_x = statistics.mean(list_x)
mean_y = statistics.mean(list_y)

covariance = sum((x - mean_x) * (y - mean_y) for x, y in zip(list_x, list_y)) / (len(list_x) - 1)

std_x = statistics.stdev(list_x)
std_y = statistics.stdev(list_y)

correlation = covariance / (std_x * std_y)

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
''' 7. Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] '''

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

x = np.array(data)

Q1 = np.percentile(x, 25, method="midpoint")
Q3 = np.percentile(x, 75, method="midpoint")
IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

outliers = x[(x < lower_fence) | (x > upper_fence)]

print(f"Q1 = {Q1}")
print(f"Q3 = {Q3}")
print(f"IQR = {IQR}")
print(f"Lower fence = {lower_fence}")
print(f"Upper fence = {upper_fence}")
print(f"Outliers = {outliers.tolist()}")

plt.boxplot(x, vert=True, showfliers=True)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
''' 8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.

1. Explain how you would use covariance and correlation to explore this
relationship.

2. Write Python code to compute the correlation between the two lists:

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]'''

''' ANS. 1. Covariance measures how two variables change together. Formula (for population) :-

            Cov(X,Y)= ∑(Xi-X)(Yi-Y)/n
            ​
          Interpretation: i.) Positive covariance → when X increases, Y tends to increase.
                          ii.)Negative covariance → when X increases, Y tends to decrease.
                          iii.)Magnitude is not standardized, so it’s hard to interpret in terms of strength.

            Correlation (Pearson’s correlation coefficient) standardizes covariance: r=Cov(X,Y)/σXσY
	​

           Range: -1 to 1.'''



import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

cov_matrix = np.cov(advertising_spend, daily_sales)
covariance = cov_matrix[0, 1]
print(f"Covariance: {covariance:.2f}")

correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]
print(f"Correlation coefficient: {correlation:.2f}")





```

```python
''' 9.  Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.

1. Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.

2. Write Python code to create a histogram using Matplotlib for the survey data:

  survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]'''

'''Ans. 1. Summary statistics are numerical measures that describe the central tendency, spread, and overall pattern of a dataset.
           The mean represents the average value, the median indicates the middle score when the data is ordered, and the mode
           shows the most frequently occurring score. The standard deviation measures how much the values deviate from the mean,
           while the minimum and maximum indicate the range of responses.Quartiles and the interquartile range divide the data into
           segments and help identify variability and outliers.Visualizations such as histograms display the frequency distribution
           of scores, boxplots summarize the median, quartiles, and outliers, and bar charts show the counts of each discrete score
           to provide a clear picture of how customer satisfaction is distributed.'''


import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=range(1, 12), edgecolor='black', align='left')
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Survey Score")
plt.ylabel("Number of Customers")
plt.xticks(range(1, 11))
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()

```
